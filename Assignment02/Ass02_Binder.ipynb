### Random Word Swap

**AI Use:** Chat GPT  
*Prompt:* give me an outline/idea for a random word swap function

```python
import random

def random_swap(sentence, swap_ratio=0.2):
    words = sentence.split()
    n_swaps = max(1, int(len(words) * swap_ratio))

    indices = list(range(len(words)))
    random.shuffle(indices)

    used = set()

    for i in indices:
        if len(used) >= n_swaps:
            break
        j = random.choice(indices)

        if i not in used and j not in used and i != j:
            words[i], words[j] = words[j], words[i]
            used.add(i)
            used.add(j)
    print("Original:")
    print(sentence)
    print("Augmented Texts:")
    print(" ".join(words))

print("Sentence 1:")
random_swap(sentence_1)
print("Sentence 2:")
random_swap(sentence_2)
print("Sentence 3:")
random_swap(sentence_3)
print("Sentence 4:")
random_swap(sentence_4)
print("sentence_5")
random_swap(sentence_5)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, AutoConfig
def roundtrip_translate(sentences, model_a_name, model_b_name):
    # Load models + tokenizers
    tokenizer_a = AutoTokenizer.from_pretrained(model_a_name)
    model_a = AutoModelForSeq2SeqLM.from_pretrained(model_a_name)
    tokenizer_b = AutoTokenizer.from_pretrained(model_b_name)
    model_b = AutoModelForSeq2SeqLM.from_pretrained(model_b_name)
    results = []
    for text in sentences:
        inputs_a = tokenizer_a(text, return_tensors="pt")
        outputs_a = model_a.generate(**inputs_a)
        translated = tokenizer_a.decode(outputs_a[0], skip_special_tokens=True)
        inputs_b = tokenizer_b(translated, return_tensors="pt")
        outputs_b = model_b.generate(**inputs_b)
        back_translated = tokenizer_b.decode(outputs_b[0], skip_special_tokens=True)
        results.append((translated,back_translated))
    return results
```

### Imports

```python
import os

import numpy as np

import nltk
import spacy
from transformers import AutoTokenizer

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
import time
```

```python
def load_imdb_folder(split="train", data_dir="aclImdb"):
    texts, labels = [], []
    for label, sentiment in [(1, "pos"), (0, "neg")]:
        folder = os.path.join(data_dir, split, sentiment)
        for fname in os.listdir(folder):
            if fname.endswith(".txt"):
                with open(os.path.join(folder, fname), encoding="utf-8") as f:
                    texts.append(f.read())
                    labels.append(label)
    return texts, labels

X_train, y_train = load_imdb_folder("train", data_dir="aclImdb")
X_test, y_test = load_imdb_folder("test", data_dir="aclImdb")

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
```

```python
X_train = np.array(X_train)
X_test  = np.array(X_test)
```

```python
# Option 1: shuffle everything (recommended)
indices = np.random.permutation(len(X_train))
X_train = X_train[indices]
y_train = np.array(y_train)[indices]

indices_test = np.random.permutation(len(X_test))
X_test = X_test[indices_test]
y_test = np.array(y_test)[indices_test]

# Now the small subsets will have both classes
X_train_small = X_train[:2000]
y_train_small = y_train[:2000]
X_test_small  = X_test[:500]
y_test_small  = y_test[:500]
```

**AI Use:** Claude Sonnet 4.6

```python
nltk.download("punkt_tab")
nlp = spacy.load("en_core_web_sm")
gpt2_tokenizer = AutoTokenizer.from_pretrained("gpt2")

def nltk_tokenizer(text):
    return nltk.word_tokenize(text.lower())

def spacy_tokenizer(text):
    return [token.text.lower() for token in nlp(str(text))]

def gpt2_tokenize(text):
    return gpt2_tokenizer.convert_ids_to_tokens(
        gpt2_tokenizer.encode(text)
    )
```

**AI Use:** Claude Sonnet 4.6  
*Prompt:* write me a for loop to make the shortest possible function to evaluate the following
- Report and compare:
    - Accuracy and F1-score on test data
    - Training time (fit time)
    - Inference time (per 1000 samples)
    - Vocabulary size from the TF–IDF layer
    - Average tokens per document for each tokenizer

```python
def build_pipeline(tokenizer_fn):
    return Pipeline([
        ("tfidf", TfidfVectorizer(
            tokenizer=tokenizer_fn,
            lowercase=True,
            token_pattern=None
        )),
        ("clf", LogisticRegression(max_iter=200, solver="lbfgs"))
    ])

tokenizers = {
    "NLTK": nltk_tokenizer,
    "spaCy": spacy_tokenizer,
    "GPT-2 BPE": gpt2_tokenize,
}

for name, tok_fn in tokenizers.items():
    print(f"\n--- {name} ---")
    pipeline = build_pipeline(tok_fn)

    # Training time
    t0 = time.time()
    pipeline.fit(X_train_small, y_train_small)
    fit_time = time.time() - t0
    print(f"Training time: {fit_time:.2f}s")

    # Inference time per 1000 samples
    sample = X_test_small[:500]
    t1 = time.time()
    pipeline.predict(sample)
    inf_time = time.time() - t1
    print(f"Inference time (1000 samples): {inf_time:.2f}s")

    # Accuracy + F1
    y_pred = pipeline.predict(X_test_small)
    print(f"Accuracy: {accuracy_score(y_test_small, y_pred):.4f}")
    print(f"F1 Score: {f1_score(y_test_small, y_pred, average='weighted'):.4f}")

    # Vocabulary size
    vocab_size = len(pipeline.named_steps["tfidf"].vocabulary_)
    print(f"Vocabulary size: {vocab_size}")

    # Average tokens per document
    avg_tokens = sum(len(tok_fn(t)) for t in X_train_small) / len(X_train_small)
    print(f"Avg tokens/doc: {avg_tokens:.1f}")
```
