### Example Webpage: 
https://arstechnica.com/science/2026/03/the-strange-animals-that-control-their-body-heat/

```python
with open(r"The strange animals that control their body heat.html", encoding="utf-8") as f:
    ars = f.read()
```

```python
ars
```

```python
ars_soup = BeautifulSoup(ars)
```

```python
article = ars_soup.select_one("article")
```

```python
article
```

```python
#Title of the article
article.select_one("h1").get_text(strip=True)
```

```python
#Content
article.select_one("div.post-content")
```

```python
# to select all the paragraphs in the content
article.select("div.post-content p")
```

```python
# to get the text of all the paragraphs in the content
text = " ".join([x.get_text(strip=True) for x in article.select("div.post-content p")])
text
```

```python
# extracting the date of publication
article.select_one("header time").get_text(strip=True) # header added to get the right publishing date and not any other date that might be in the content
```

```python
# extracting the update date
article.select_one("time[datetime]")["datetime"]
```

```python
import requests
from urllib.robotparser import RobotFileParser
```

```python
response = requests.get("https://arstechnica.com/science/2026/03/the-strange-animals-that-control-their-body-heat/")
response
```

```python
rp = RobotFileParser("https://arstechnica.com/science/2026/03/the-strange-animals-that-control-their-body-heat/")
rp.read()
```

```python
rp.can_fetch("NLP ExerciseScraper", "https://arstechnica.com/science/2026/03/the-strange-animals-that-control-their-body-heat/")
```

```python
rp.site_maps
```

```python
# for rate limiting 
rp.request_rate("NLP ExerciseScraper")
```
