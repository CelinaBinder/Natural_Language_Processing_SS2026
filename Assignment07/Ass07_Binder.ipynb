# Exercise 2 LangChain Prompt & Chat Templates: Domain-Specific Q&A Assistant
Large Language Models can be adapted to specialized domains by carefully designing prompts
and chat templates. In this exercise, you will build a reusable, domain-specific Q&A assistant
using LangChain prompt templates instead of ad-hoc prompts.  
The assistant should answer questions based on provided policy or guideline excerpts and produce
answers in a consistent, structured format.
Tasks:  
1. Design a LangChain PromptTemplate or chat prompt template that accepts at least the  
following inputs:  
• a domain document or policy excerpt,  
• a user question,  
• a parameter controlling the level of formality (e.g., casual, professional, legal).  
2. Enforce a fixed response structure (e.g., short answer, justification, and reference to the
provided text).  
3. Test your template with multiple domain documents and questions.  
4. Briefly discuss the advantages of prompt templates over inline prompting in larger applications.  
  
Deliverables: A notebook containing the prompt template implementation, example interactions,
and a short discussion of design decisions.  

### 1. Design LangChain Prompt Template
and
### 2. enforce fixed response structure

*Source:* https://langchain-tutorials.com/lessons/langchain-essentials/lesson-6

````python
# Task 1: Define formality levels and instructions
# level of formality
FORMALITY = {
    "casual":       "use simple, everyday language",
    "professional": "use clear, professional language suitable for a business context",
}

# Template and fixed response structure
template_string = """You are a domain specific Q&A assistant.
Answer the question strictly based on the provided domain document or policy excerpt. Do not use outside knowledge.

Tone & style: {formality_instruction}

Domain document/Policy excerpt:
{document}

Question: {question}

You MUST always respond in exactly this structure:

Short Answer: <one or two sentence direct answer>

Justification: <explain your reasoning based on the document>

Reference: <quote the specific part of the document supporting your answer>
"""

# Task 2: fixed structure is enforced inside the template string above
prompt = PromptTemplate(
    template=template_string,
    input_variables=["formality_instruction", "document", "question"]
)
````

### 3. Test your template

**AI Use:** Claude Sonnet 4.6  
*Prompt:* give me three examples of domain documents or policy excerpts and questions to ask about them to use for my exercise *Exercise Sheet*

````python
# Example 1: HR Policy, professional
print("Example 1: HR Policy, professional")
hr_policy = """
Employees are entitled to 25 days of paid annual leave per calendar year.
Leave must be requested at least two weeks in advance via the HR portal.
Unused leave of up to 5 days may be carried over to the following year.
Carry-over must be approved by the employee's line manager.
"""

formatted_prompt = prompt.format(
    document=hr_policy,
    question="Can I carry over unused leave to next year?",
    formality_instruction=FORMALITY["professional"]
)
print(llm.invoke(formatted_prompt).content)
print("-"*100)

# Example 2: Food Safety Guidelines, casual
print("Example 2: Food Safety Guidelines, casual")
food_safety = """
All food handlers must wash their hands before touching any food.
Raw meat must be stored on the lowest shelf of the refrigerator.
Cooked food must not be left at room temperature for more than 2 hours.
All surfaces must be sanitized after contact with raw meat or fish.
"""

formatted_prompt = prompt.format(
    document=food_safety,
    question="Where should I store raw chicken in the fridge?",
    formality_instruction=FORMALITY["casual"]
)
print(llm.invoke(formatted_prompt).content)
print("-"*100)

# Example 3: GDPR Excerpt, legal
print("Example 3: GDPR Excerpt, legal")
gdpr_excerpt = """
Personal data shall be collected for specified, explicit and legitimate purposes
and not further processed in a manner that is incompatible with those purposes.
Data subjects have the right to request erasure of their personal data without
undue delay where the data is no longer necessary for the purpose it was collected.
"""

formatted_prompt = prompt.format(
    document=gdpr_excerpt,
    question="Can a user request their data to be deleted?",
    formality_instruction=FORMALITY["professional"]
)
print(llm.invoke(formatted_prompt).content)
print("-"*100)
````

### 2. Langchain Memory

**AI Use:** Microsof CoPilot Smart

````python
store = {}
def get_history(thread_id: str):
    if thread_id not in store:
        store[thread_id] = InMemoryChatMessageHistory()
    return store[thread_id]
````

````python
def interview_logic(inputs):
    formatted = interview_prompt.format_messages(  
        history=inputs["history"],
        input=inputs["input"]
    )
    return llm.invoke(formatted)
````

````python
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.runnables import ConfigurableFieldSpec

interview = RunnableWithMessageHistory(
    interview_runnable,
    get_history,
    input_messages_key="input",
    history_messages_key="history",
    history_factory_config=[                        
        ConfigurableFieldSpec(
            id="thread_id",                         
            annotation=str,
            name="Thread ID",
            default="",
        )
    ]
)
````

### 3. Add Checkpointing

*Source:* https://docs.langchain.com/oss/python/langgraph/persistence

````python
from langgraph.checkpoint.memory import MemorySaver
from typing import TypedDict
from langgraph.graph import StateGraph, END
````

````python
class InterviewState(TypedDict):
    input: str
    session_id: str
    response: str | None

checkpointer = MemorySaver()

def interview_node(state: InterviewState):
    result = interview.invoke(
        {"input": state["input"]},
        config={"configurable": {"thread_id": state["session_id"]}}
    )
    return {"response": result.content}

graph = StateGraph(InterviewState)
graph.add_node("interview", interview_node)
graph.set_entry_point("interview")
graph.add_edge("interview", END)
app = graph.compile(checkpointer=checkpointer)
````

### 4. Demonstrate that the interaction can be stopped and resumed

````python
session_id = "candidate_1"

result = app.invoke(
    {"input": "I'm ready to start.", "session_id": session_id, "response": None},
    config={"configurable": {"thread_id": session_id}}
)
print(result["response"])

result = app.invoke(
    {"input": "I have no Python experience.", "session_id": session_id, "response": None},
    config={"configurable": {"thread_id": session_id}}
)
print(result["response"])
````

### 5. Compare at least two different memory strategies (e.g., full buffer vs. summarized memory).  
*Source:* https://www.geeksforgeeks.org/artificial-intelligence/memory-in-langchain-1/    
  
**Entity Memory:**  
- extracts and remembers structured facts about specific entities
- enables personalized interactions 
    - captures and updates entity-specific data
    - supports personalized assistants and domain knowledge
    - stored as structured representations for easy retrieval

  
**Full Buffer:**  
- stores entire conversation history as it occurred and sequentially correct
- can be inefficient with long conversations because of token limits 
    - stores full message history
    - great for short conversations or prototyping
    - easy to implement but unbounded growth can affect performance

**AI Use:** Claude Sonnet 4.6  
*Prompt:* for the described memory strategies compare them with two short code snippets

**Full Buffer**

````python
# Strategy 1: Full Buffer
session_id_buffer = "buffer_candidate"

app.invoke(
    {"input": "I have 2 years of Python experience.", "session_id": session_id_buffer, "response": None},
    config={"configurable": {"thread_id": session_id_buffer}}
)
app.invoke(
    {"input": "I worked as a backend developer at a fintech company.", "session_id": session_id_buffer, "response": None},
    config={"configurable": {"thread_id": session_id_buffer}}
)

print("---- Full Buffer - complete history ----")
for msg in store.get(session_id_buffer).messages:
    role = "Candidate" if msg.type == "human" else "Interviewer"
    print(f"{role}: {msg.content}\n")
````

**Entitiy Memory**

````python
# Strategy 2: Entity Memory
import json
from langchain_core.messages import SystemMessage, HumanMessage

candidate_profile = {}

def extract_entities(answer: str) -> dict:
    response = llm.invoke(
        f"""Extract candidate facts as flat JSON.
Only include fields explicitly mentioned: name, skills, experience_years, previous_role, company.
Answer: "{answer}"
Respond with JSON only."""
    )
    try:
        return json.loads(response.content)
    except:
        return {}

answers = [
    "I'm Sara, I have 4 years of Java experience.",
    "I worked as a backend developer at a fintech company.",
]

for answer in answers:
    candidate_profile.update(extract_entities(answer))
    print(f"[Profile updated]: {candidate_profile}")

    messages = [
        SystemMessage(content=f"You are an HR interviewer. Known facts about candidate: {candidate_profile}"),
        HumanMessage(content=answer)
    ]
    response = llm.invoke(messages)
    print(f"Interviewer: {response.content}\n")

print("=== Entity Memory — structured profile only ===")
print(json.dumps(candidate_profile, indent=2))
````
