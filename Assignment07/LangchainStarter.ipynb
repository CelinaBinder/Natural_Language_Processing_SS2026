
````python
from langchain.messages import HumanMessage
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOllama(
    model="ministral-3:14b",
    temperature=0.2,
    streaming=True,
    verbose=True
)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Answer questions as concisely as possible."),
    ("human", "{question}")
])

chain = prompt | llm

for chunk in chain.stream({"question": "Tell me about langchain."}):
    if chunk.content:
        print(chunk.content, end="", flush=True)
````
